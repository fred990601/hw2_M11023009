# #資料量化

```python
def dataCleaning(dataSet):
    #量化_換成number 0 1 2 3 (量化)
    beanclass = list(set(dataSet['Class']))
    print(beanclass)
    for i in range(len(beanclass)):
        row = dataSet['Class'] == beanclass[i]
        dataSet.loc[row,'Class'] = i
    dataSet['Class'] = dataSet['Class'].astype(np.int64)
      
    return dataSet
```

```python
Dry_Bean_Dataset = dataCleaning(Dry_Bean_Dataset)
Dry_Bean_Dataset
```

```python
Dry_Bean_Dataset = Dry_Bean_Dataset.sample(frac=1)
Dry_Bean_Dataset
```
